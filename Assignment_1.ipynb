```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

employee_rating = pd.read_csv('employee_rating.csv')
employee_general = pd.read_csv('employee_general.csv')
print("Employee Rating Data:")
display(employee_rating.head())
print("\nEmployee General Data:")
display(employee_general.head())
print("\nData types for employee_rating:")
display(employee_rating.dtypes)
print("\nData types for employee_general:")
display(employee_general.dtypes)
```

# **2. Generate a log_income column with log_income = log(MonthlyIncome) for the employee_rating dataset**

```python
import numpy as np

print("\nSkewness of MonthlyIncome:", employee_rating['MonthlyIncome'].skew())
employee_rating['log_income'] = np.log(employee_rating['MonthlyIncome'])

print("\nEmployee Rating Data with log_income column:")
display(employee_rating[['MonthlyIncome', 'log_income']].head())
```

# **3. Convert class columns to meaningful numbers in the employee_rating dataset**

```python
# Step 1: Print the actual column names to identify class columns that can be converted
print("\nColumns in employee_rating dataset:")
print(employee_rating.columns)

# Step 2: Convert the 'EnvironmentSatisfaction' and 'PerformanceRating' columns into numerical values
# Replace these with actual categories you want to convert.

# Example: Convert 'EnvironmentSatisfaction' to numerical values (Low=1, Medium=2, High=3)
employee_rating['EnvironmentSatisfaction'] = employee_rating['EnvironmentSatisfaction'].map({'Low': 1, 'Medium': 2, 'High': 3})

# Convert 'PerformanceRating' to numerical values (Below Expectations=1, Meets Expectations=2, Exceeds Expectations=3)
employee_rating['PerformanceRating'] = employee_rating['PerformanceRating'].map({'Below Expectations': 1, 'Meets Expectations': 2, 'Exceeds Expectations': 3})

# Step 3: Display the updated dataset to verify the changes
print("\nUpdated employee_rating dataset:")
display(employee_rating.head())
```

# **4. Convert class columns to meaningful numbers in the employee_general dataset**

```python
employee_general = pd.read_csv('employee_general.csv', delimiter='|')

print("\nColumns in employee_general dataset:")
print(employee_general.columns)

employee_general['MaritalStatus'] = employee_general['MaritalStatus'].map({'Single': 1, 'Married': 2, 'Divorced': 3})

employee_general['Education'] = employee_general['Education'].map({'Undergraduate': 1, 'Graduate': 2, 'Postgraduate': 3})

print("\nUpdated employee_general dataset:")
display(employee_general.head())
```

# **5. Merge the two datasets**

```python
# Step 12: Merge the datasets on the common column, which is likely 'EmployeeNumber'
merged_data = pd.merge(employee_rating, employee_general, on='EmployeeNumber', how='inner')

# Step 13: Display the merged dataset
print("\nMerged Dataset:")
display(merged_data.head())
```
